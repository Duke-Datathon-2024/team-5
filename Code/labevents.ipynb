```python
import pandas as pd
import numpy as np
```

```python
emar = pd.read_csv("emar_emar_detail_vancomycin_IV.csv")
```

```python
irrelevant = [
    "prior_infusion_rate",
    "infusion_rate",
    "infusion_rate_adjustment",
    "infusion_rate_adjustment_amount",
    "infusion_rate_unit",
    "route",
    "infusion_complete",
    "completion_interval",
    "new_iv_bag_hung",
    "continued_infusion_in_other_location",
    "restart_interval",
    "side",
    "site",
    "non_formulary_visual_verification",
    "parent_field_ordinal",
    "dose_due",
    "complete_dose_not_given",
    "dose_due_unit",
    "product_description_other",
    "will_remainder_of_dose_be_given",
    "reason_for_no_barcode",
    "subject_id_1",
    "emar_id_1",
    "emar_seq_1",
    "administration_type",
    "pharmacy_id_1",
]
```

```python
# remove the rows where product description is 'vancomycin', 'vancomycin 125 mg capsule', 'vancomycin 125mg','vancomycin 125 mg', 'vancomycin 250 mg capsule'

emar = emar[
    ~emar["product_description"].isin(
        [
            "vancomycin",
            "vancomycin 125 mg capsule",
            "vancomycin 125mg",
            "vancomycin 125 mg",
            "vancomycin 250 mg capsule",
        ]
    )
]

# remove irrelevant columns
emar = emar.drop(columns=irrelevant)
```

```python
# remove rows where dose_given is null
emar = emar[emar["dose_given"].notnull()]
emar[emar["subject_id"] == 18522065]
```

```python
labevents = pd.read_csv("labevents_51009_vancomycin.csv")
```

```python
labevents
```

```python
# merge emar and labevents based on subject_id and if charttime for emar is within 48 hours of charttime for labevents

emar["charttime"] = pd.to_datetime(emar["charttime"])
labevents["charttime"] = pd.to_datetime(labevents["charttime"])

emar["charttime"] = pd.to_datetime(emar["charttime"])

labevents["charttime"] = pd.to_datetime(labevents["charttime"])

emar_labevents = pd.merge(emar, labevents, on="subject_id", how="inner")

emar_labevents["diff"] = emar_labevents["charttime_x"] - emar_labevents["charttime_y"]

emar_labevents["diff"] = emar_labevents["diff"].dt.total_seconds()

emar_labevents["diff"] = emar_labevents["diff"] / 3600

emar_labevents = emar_labevents[
    (emar_labevents["diff"] >= -48) & (emar_labevents["diff"] <= 0)
]

emar_labevents[emar_labevents["subject_id"] == 18522065]
```

```python
selected_columns = emar_labevents[
    [
        "subject_id",
        "hadm_id_x",
        "emar_id",
        "charttime_x",
        "charttime_y",
        "diff",
        "dose_given",
        "dose_given_unit",
        "valuenum",
        "value",
    ]
]
```
